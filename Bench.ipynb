## Benchamrks against BLAS

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_float = pd.read_csv("./float.csv", sep = ";")
data_double = pd.read_csv("./double.csv", sep = ";")
```

## Plots

```python
plt.plot(data_float["size"], data_float["matrice"], color = "black", label = "matrice")
plt.plot(data_float["size"], data_float["blas"], color = "red", label = "blas")
plt.legend()
plt.show()
```

```python
plt.plot(data_double["size"], data_double["matrice"], color = "black", label = "matrice")
plt.plot(data_double["size"], data_double["blas"], color = "red", label = "blas")
plt.legend()
plt.show()
```
